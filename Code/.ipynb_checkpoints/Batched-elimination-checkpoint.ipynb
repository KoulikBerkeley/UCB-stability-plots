```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python
import numpy as np

class BatchedBandit:
    def __init__(self, true_means, confidence_multiplier = 1.0):
        self.true_means = true_means
        self.num_arms = len(self.true_means)
        self.confidence_multiplier = confidence_multiplier
        self.total_rewards = np.zeros(self.num_arms)
        self.total_pulls = np.zeros(self.num_arms)
        self.is_active = np.ones(self.num_arms, dtype=bool)

    def choose_arm_and_update(self):
        active_arms = np.where(self.is_active)[0]
        if len(active_arms) == 0:
            return  # No active arms left
        arm_index = np.random.choice(active_arms)
        reward = np.random.normal(self.true_means[arm_index], 1)  
        self.total_rewards[arm_index] += reward
        self.total_pulls[arm_index] += 1


    def eliminate_arms(self):
        upper_confidence_bounds = self.total_rewards / (self.total_pulls + 1e-9) + \
                                  self.confidence_multiplier * np.sqrt(np.log(np.sum(self.total_pulls)) / (self.total_pulls + 1e-9))
        for i in range(self.num_arms):
            if self.is_active[i]:
                for j in range(self.num_arms):
                    if i != j and self.is_active[j]:
                        lower_confidence_bound_j = self.total_rewards[j] / (self.total_pulls[j] + 1e-9) - \
                                                   self.confidence_multiplier * np.sqrt(np.log(np.sum(self.total_pulls)) / (self.total_pulls[j] + 1e-9))
                        if upper_confidence_bounds[i] < lower_confidence_bound_j:
                            self.is_active[i] = False
                            break
                            
                            
    def sample_average_algorithm(self):
        sample_averages = np.zeros(self.num_arms)
        for arm_index in range(self.num_arms):
            if self.total_pulls[arm_index] == 0:
                sample_averages[arm_index] = 0
            else:
                sample_averages[arm_index] = self.total_rewards[arm_index] / self.total_pulls[arm_index]
        return sample_averages
    
    
    def difference_confidence_bounds(self, arm_index_1, arm_index_2):
        sample_averages = self.sample_average_algorithm()
        n_1 = self.total_pulls[arm_index_1]
        n_2 = self.total_pulls[arm_index_2]
        if n_1 == 0 or n_2 == 0:
            return float('inf')
        else:
            std_dev_1 = 1 / np.sqrt(n_1)
            std_dev_2 = 1 / np.sqrt(n_2)
            
            diff_mean = sample_averages[arm_index_1] - sample_averages[arm_index_2]
            true_mean_diff = self.true_means[arm_index_1] - self.true_means[arm_index_2] 
            margin_of_error = (diff_mean - true_mean_diff) / np.sqrt(std_dev_1**2 + std_dev_2**2)
            
            return margin_of_error
    


def get_margin_zscore_multi_round_eps_ucb(true_means, samples_per_round, arm_index_1, 
                                      arm_index_2, T):
    bandit = BatchedBandit(true_means)

    num_pull_per_arm_first_round = samples_per_round // bandit.num_arms
    

    # Pull each arm equally often in the first n rounds
    for arm_index in range(bandit.num_arms):
        for _ in range(num_pull_per_arm_first_round):
            bandit.choose_arm_and_update()
            
            
    # Running elimination at the end of first round.         
    bandit.eliminate_arms()
            
    
    
    # For the remaining rounds pull all remaining arms at random and 
    # run elimination after every round.
    
    for _ in range(T - 1):  # Already performed 1 stage in the first round
        # Epsilon-greedy step
        
        for _ in range(samples_per_round):
            bandit.choose_arm_and_update()
        
        # Running elimination after every round.         
        bandit.eliminate_arms()
            
        
    return bandit.total_rewards, bandit.total_pulls, bandit.difference_confidence_bounds(arm_index_1, arm_index_2)
```

```python
true_means = [0.0, 0.0]  # True means
num_arms = len(true_means)

samples_per_round = 25
num_rounds = 25
arm_index_1 = 0
arm_index_2 = 1

get_margin_zscore_multi_round_eps_ucb(true_means, samples_per_round, arm_index_1, 
                                      arm_index_2, num_rounds)
```

```python
# Define the parameters for the experiment
num_experiments = 10000
true_means = [0.0, 0.0, -0.5]  # True means
num_arms = len(true_means)

samples_per_round = 25
num_rounds = 25
arm_index_1 = 0
arm_index_2 = 1



# Perform the experiments
margin_zscores = []
average_pulls_data = np.zeros((num_experiments, len(true_means)))
average_reward_data = np.zeros((num_experiments, len(true_means)))

for exp_idx in range(num_experiments):
    total_reward, arm_pulls, zscores = get_margin_zscore_multi_round_eps_ucb(true_means, samples_per_round, arm_index_1, arm_index_2,
                                            num_rounds)
    margin_zscores.append(zscores)
    average_pulls_data[exp_idx] = arm_pulls
    average_reward_data[exp_idx] = total_reward



plt.figure(figsize=(10, 6))

sns.histplot(margin_zscores, kde=True, stat="density",  label='margin')
plt.plot(np.linspace(-4, 4, 1000), stats.norm.pdf(np.linspace(-4, 4, 1000)), color='red', label='Standard Normal Density')

# Get the average number of arm pulls for each arm
average_pulls = np.mean(average_pulls_data, axis=0)

# Create labels with the average number of arm pulls
arm_labels = [f'Arm {arm_index} (Avg Pulls: {average_pulls[arm_index]:.2f})' for arm_index in range(num_arms)]

plt.title('Density of Confidence Bounds for Each Arm vs Standard Normal Density')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend(labels=arm_labels)
plt.grid(True)
plt.show()

_ = stats.probplot(margin_zscores, dist="norm", plot=plt)

```

```python
alpha = 0.05

# # Fit a normal distribution to the data
mean, std_dev = np.mean(margin_zscores), np.std(margin_zscores)
normal_distribution = stats.norm(loc=mean, scale=std_dev)
# Perform the Kolmogorov-Smirnov test
ks_statistic, p_value = stats.kstest(margin_zscores, normal_distribution.cdf)
print(f'KS Statistic = {ks_statistic}, p-value = {p_value}')
if p_value > alpha:
    print("Null hypothesis (data follows a normal distribution) cannot be rejected.")
else:
    print("Null hypothesis (data follows a normal distribution) is rejected.")
```
